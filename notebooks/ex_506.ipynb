```python
#lemmatization -> reducing to word to its root form based on dictonary meaning. It is a more accurate process and gives the correct root word. not by internal rules
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
from nltk.corpus import wordnet
lemmatizer = WordNetLemmatizer()
```

```python
import nltk
import re
```

```python
sentences = nltk.sent_tokenize(paragraph)
for i in range(len(sentences)):
    sentences[i] = sentences[i].lower()
    sentences[i] = re.sub(r'\W',' ',sentences[i])
    sentences[i] = re.sub(r'\s+',' ', sentences[i])
    sentences[i] = re.sub(r's$', ' ', sentences[i])
```

```python
sentences
```

```python
#stop words : are common function words that carry little semantic value and are often removed from text during preprocessing 
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [word for word in words if word not in stop_words]
    sentences[i] = ' '.join(words)
sentences
```

```python
#NN = noun 
#VB = verb
#CJ = adjective
#RB = adverb
my_sentence = "The quick brown fox jumps over the lazy dog"
```

```python
my_words = nltk.word_tokenize(my_sentence)
for i in range(len(my_words)):
    pos = nltk.pos_tag([my_words[i]])[0][1]
    if pos.startswith('NN'):
        my_words[i] = lemmatizer.lemmatize(my_words[i], pos=wordnet.NOUN)
    elif pos.startswith('VB'):
        my_words[i] = lemmatizer.lemmatize(my_words[i], pos=wordnet.VERB)
    elif pos.startswith('JJ'):
        my_words[i] = lemmatizer.lemmatize(my_words[i], pos=wordnet.ADJ)
    elif pos.startswith('RB'):
        my_words[i] = lemmatizer.lemmatize(my_words[i], pos=wordnet.ADV)
```

```python
my_words = nltk.word_tokenize(my_sentence)
my_tags = nltk.pos_tag(my_words)
my_tags
```

```python
my_words_tagged = []
for tw in my_words:
    pos = nltk.pos_tag([tw])[0][1]
    print(f"{tw} : {pos}")
    for tw in my_words:
        my_words_tagged.append(tw[0] + '(' + pos + ')')
my_sentence_tagged = ' '.join(my_words_tagged)
my_sentence_tagged
```
