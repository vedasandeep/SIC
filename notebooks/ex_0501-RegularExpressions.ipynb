Meta Character: {m,n} 

{m,n} means the preceding character must appear at least m times and at most n times

```python
x4 = "caaat"
my_expression = re.compile("ca{1,2}t")
if my_expression.search(x1):
    print("Yes")
else:
    print("No")
```

Meta Character: ^ (anchor)
match the start of string

```python
x = ["This is my book","That is your book"]
my_exp = re.compile("^This")
for i in x:
    if my_exp.search(i):
        print(i,": Yes")
    else:
        print(i,": No")
```

^[] Matches the start of string, either upper case or lower case


```python
x = ["I dont like you","You are great","in case you forget,remember this"]
my_exp = re.compile("^[Ii]")
for i in x:
    if my_exp.search(i):
        print(i,": Yes")
    else:
        print(i,": No")
```

^ anchor with digit class

```python
x = ["2This is my book","1That is your book"]
my_exp = "^[0-9]"
for i in x:
    if re.search(my_exp,i):
        print(i,": Yes")
    else:
        print(i,": No")
```

```python
x = ["This is year 2019","2000 is the year of my birth"]
my_exp = "^[0-9a-zA-Z]"
for i in x:
    if re.search(my_exp,i):
        print("Yes")
    else:
        print("No")
```

$ : end of the string

```python
x = ["Python is easy","You need python"]
my_exp = "n$"
for i in x:
    if re.search(my_exp,i):
        print(i,": Yes")
    else:
        print(i,": No")
```

Meta Character : |
left or right pattern

```python
x = ["I Love You","I Love You","I Love You","I Love You"]
my_exp = "love|hate"
for i in x:
    if re.search(my_exp,i):
        print("Yes")
    else:
        print("No")
```

```python
def grep(reg_exp, my_text):
    result = [] #creates a list to store matching lines  # noqa: F841
    reg_exp_compiled = re.compile(reg_exp)  # noqa: F841
    
    for i in range(len(my_text)):
        if reg_exp_compiled.search(my_text[i]):
            result.append(my_text[i])
    return result
```

```python
reg_ex = re.compile("([0-9]+)")
m = reg_ex.search("I am 22 yeas old, she is 20 years")
print(m)
```

```python
reg_ex = re.compile("([0-9]+)[^0-9]+([0-9]+)")
m = reg_ex.search("I am 22 years old. She is 20 years")
print(m.group(1))
print(m.group(2))
print(m)
```

```python
reg_ex = re.compile("([^0-9]+)([0-9]+)[^0-9]+([0-9]+)[^0-9]+([0-9]+)")
m = reg_ex.search("John Wheeler 010-1234-5678")
print((m.group(1)).strip())
print(m.group(2))
print(m.group(3))
print(m.group(4))
```

# short hand notations:
# \d = [0-9]
# \D = [^0-9]
# \w = [0-9a-zA-Z]
# \W = [^0-9a-zA-Z_]    

```python
# short hand notations:
# \d = [0-9]
# \D = [^0-9]
# \w = [0-9a-zA-Z]
# \W = [^0-9a-zA-Z_]    
```

```python
reg_ex = re.compile("(\d+)")
m = reg_ex.search("I am 22 year old, She is 20 years old")
print(m.group())
```

```python
for i in range(1,m.lastindex+1):
    value = m.group(i)
    if i == 1:
        print(value.strip())
    else:
        print(value)
```

```python
reg_ex = re.compile("(\D+)((\d+)\D+(\d+)\D+(\d+))")
m = reg_ex.search("John Wheeler 010-1234-5678")
print("PHONE : "+m.group(2))
```
