### Integer Encoding in NLP is a method that maps each unique token to a unique integer ID.

### A vocabulary is built from the text. Each word is replaced by its index in that vocabulary

```python
from nltk.tokenize import sent_tokenize  # noqa: F401
from nltk.tokenize import word_tokenize # noqa: F401
from nltk.corpus import stopwords # noqa: F401
```

```python
text = "A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
```

```python
#sentence tokenization
text = sent_tokenize(text)
print(text)
```

```python
vocab = {}
sentences = [] 
stop_words = set(stopwords.words('english'))
 
for i in text:
    sentence = word_tokenize(i)
    result = []
    
    for word in sentence:
        word = word.lower()
        if word not in stop_words:
            if len(word) > 2:
                result.append(word)
                if word not in vocab:
                    vocab[word] = 0
                vocab[word] += 1
    sentences.append(result)
print(sentences)
```

```python
vocab 
```

```python
vocab_sorted = sorted(vocab.items(), key = lambda x: x[1], reverse = True)
vocab_sorted
```

```python
word_to_index = {}
i = 0
for (word,frequency) in vocab_sorted:
    if frequency>1:
        i=i+1 
        word_to_index[word] = i
print(word_to_index)
```

```python
vocab_size = 5 

words_frequency = [w for w,c in word_to_index.items() if c >= vocab_size +1]
for w in words_frequency:
    del word_to_index[w]
print(word_to_index)
```

```python
word_to_index['OOV'] = len(word_to_index) + 1
```

```python
encoded = []
for s in sentences:
    temp = [] 
    for w in s:
        try:
            temp.append(word_to_index[w])
        except KeyError:
            temp.append(word_to_index['OOV'])
    encoded.append(temp)
print(encoded)
print(sentences)
```
