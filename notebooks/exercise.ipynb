Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
print(titanic.head())
```

```python
print(titanic.columns)
```

```python
print(titanic.info)
```

```python
titanic.describe()
```

```python
titanic.info
```

```python
print(titanic.info())
```

```python
print(titanic.describe())
```

```python
df = titanic.loc[:,['age','fare']]
print('\n First 5 rows')
print(df.head())
print('\n Last 5 rows')
print(df.tail())
print(type(df))
```

```python
import matplotlib.pyplot as plt

# Create visualizations using seaborn with the titanic dataset
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Distribution of age
sns.histplot(data=titanic, x='age', bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Age Distribution')

# Plot 2: Survival count
sns.countplot(data=titanic, x='survived', ax=axes[0, 1])
axes[0, 1].set_title('Survival Count')

# Plot 3: Survival by class
sns.countplot(data=titanic, x='pclass', hue='survived', ax=axes[1, 0])
axes[1, 0].set_title('Survival by Passenger Class')

# Plot 4: Fare distribution by class
sns.boxplot(data=titanic, x='pclass', y='fare', ax=axes[1, 1])
axes[1, 1].set_title('Fare Distribution by Class')

plt.tight_layout()
plt.show()
```
